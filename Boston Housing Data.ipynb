# Train Test Splitting


```python
x=housing.drop(['MEDV'],axis=1)
y=housing[['MEDV']]
```

#Dividing train test data in 80&20 percent ratio.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=15)
```

```python
x_train.shape
```

```python
y_train.shape
```

# selecting desired model

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

# Using Linear Regression

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
LR_pred = LR.predict(x_test)
print(r2_score(y_test,LR_pred)*100)
print(LR.score(x_test,y_test)*100)
print(LR.score(x_train,y_train)*100)
print(mean_squared_error(y_test,LR_pred))
```

# using Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
RF=RandomForestRegressor(max_depth=30)
RF.fit(x_train,y_train)
RF_pred = RF.predict(x_test)
print(r2_score(y_test,RF_pred)*100)
print(RF.score(x_train,y_train)*100)
print(RF.score(x_test,y_test)*100)
print(mean_squared_error(y_test,RF_pred))
```

# Using Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor(max_depth=10)
DT.fit(x_train,y_train)
DT_pred = DT.predict(x_test)
print(r2_score(y_test,DT_pred)*100)
print(DT.score(x_train,y_train)*100)
print(DT.score(x_test,y_test)*100)
print(mean_squared_error(y_test,DT_pred))
```

# Using KNeighbors Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train,y_train)
knn_pred = knn.predict(x_test)
print(r2_score(y_test,knn_pred)*100)
print(knn.score(x_train,y_train)*100)
print(knn.score(x_test,y_test)*100)
print(mean_squared_error(y_test,knn_pred))
```

 # Using AdaBoost Regressor

```python
from sklearn.ensemble import AdaBoostRegressor
AD = AdaBoostRegressor(n_estimators=100)
AD.fit(x_train,y_train)
AD_pred = AD.predict(x_test)
print(r2_score(y_test,AD_pred)*100)
print(AD.score(x_train,y_train)*100)
print(AD.score(x_test,y_test)*100)
print(mean_squared_error(y_test,AD_pred))
```

# Using Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
GB = GradientBoostingRegressor(n_estimators=100)
GB.fit(x_train,y_train)
gb_pred = GB.predict(x_test)
print(r2_score(y_test,gb_pred)*100)
print('train accuracy:',GB.score(x_train,y_train)*100)
print('test accuracy:',GB.score(x_test,y_test)*100)
print(mean_squared_error(y_test,gb_pred))
```

# Using XGB Regressor

```python
from xgboost import XGBRegressor
XGB = XGBRegressor()
XGB.fit(x_train,y_train)
xgb_pred = XGB.predict(x_test)
print(r2_score(y_test,xgb_pred)*100)
print('train accuracy:',XGB.score(x_train,y_train)*100)
print('test accuracy:',XGB.score(x_test,y_test)*100)
print(mean_squared_error(y_test,xgb_pred))
```

# Using LGBM Regressor

```python
import lightgbm as lgbm
LG= lgbm.LGBMRegressor()
LG.fit(x_train,y_train)
lg_pred = LG.predict(x_test)
print(r2_score(y_test,lg_pred)*100)
print('train accuracy:',LG.score(x_train,y_train))
print('test accuracy:',LG.score(x_test,y_test))
print(mean_squared_error(y_test,lg_pred))
```
